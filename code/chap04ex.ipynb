Read the pregnancy file.

```python
%matplotlib inline

import nsfg
preg = nsfg.ReadFemPreg()
```

Select live births, then make a CDF of <tt>totalwgt_lb</tt>. 

```python
import thinkstats2

live = preg[preg.outcome == 1]
cdf = thinkstats2.Cdf(live.totalwgt_lb, label='totalwgt_lb')
```

Display the CDF.

```python
import thinkplot

thinkplot.cdf(cdf)
thinkplot.Show(xlabel='totalwgt_lb', ylabel='CDF')
```

Find out how much you weighed at birth, if you can, and compute CDF(x). 

```python
tymek_weight = 8.598028218
cdf.PercentileRank(tymek_weight)
```

If you are a first child, look up your birthweight in the CDF of first children; otherwise use the CDF of other children.

```python
import first

live, firsts, others = first.MakeFrames()
firsts_cdf = thinkstats2.Cdf(firsts.totalwgt_lb, label='firsts birtweight')
thinkplot.cdf(firsts_cdf)
thinkplot.Show(xlabel='firsts totalwgt_lb', ylabel='CDF')
```

Compute the percentile rank of your birthweight

```python
firsts_cdf.PercentileRank(tymek_weight)
```

Compute the median birth weight by looking up the value associated with p=0.5.

```python
p = 0.5
cdf.Value(p)
```

Compute the interquartile range (IQR) by computing percentiles corresponding to 25 and 75. 

```python
interquartile_25 = cdf.Percentile(25)
interquartile_75 = cdf.Percentile(75)
IQR = (interquartile_25, interquartile_75)
IQR
```

Make a random selection from <tt>cdf</tt>.

```python
cdf.Random()
```

Draw a random sample from <tt>cdf</tt>.

```python
sample = cdf.Sample(1000).tolist()
```

Draw a random sample from <tt>cdf</tt>, then compute the percentile rank for each value, and plot the distribution of the percentile ranks.

```python
percentile_ranks = [cdf.PercentileRank(x) for x in sample]
rank_cdf = thinkstats2.Cdf(percentile_ranks, label='sample')
thinkplot.Cdf(rank_cdf)
thinkplot.Show(xlabel='percentile rank', ylabel='CDF')
```

Generate 1000 random values using <tt>random.random()</tt> and plot their PMF.

```python
sample_pmf = thinkstats2.Pmf(sample, label='sample pmf')
thinkplot.Pmf(sample_pmf)
thinkplot.Show(xlabel='sample', ylabel='pmf')
```

Assuming that the PMF doesn't work very well, try plotting the CDF instead.

```python
sample_cdf = thinkstats2.Cdf(sample, label='sample cdf')
thinkplot.pmf(sample_cdf)
thinkplot.Show(xlabel='sample', ylabel='CDF')
```
